```python
#@markdown 🔄 Making functions: etts('text'), ktts('text'), btts('text)
%%capture
!pip install gTTS
from gtts import gTTS
from IPython.display import Audio

def etts(text):
  text_to_say = text

  gtts_object = gTTS(text = text_to_say,
                     lang = "en", tld = "us",
                    slow = False)

  gtts_object.save("E-audio.mp3")
  return Audio("E-audio.mp3")

def ktts(text):
  text_to_say = text

  gtts_object = gTTS(text = text_to_say,
                     lang = "ko",
                    slow = False)

  gtts_object.save("K-audio.mp3")
  return Audio("K-audio.mp3")

def btts(text):
  text_to_say = text

  gtts_object = gTTS(text = text_to_say,
                     lang = "en", tld = "co.uk",
                    slow = True)

  gtts_object.save("B-audio.mp3")
  return Audio("B-audio.mp3")
```

```python
#@markdown Articulation tips and [a] vowel comparision
text = "Articulation tips: [æ] vowel is articulated with a lower jaw position. Try to make the jaw position similar to the Korean ‘ah’."
text1 = "hot, mom, father"
etts(text); ktts(text1)
print("Articulation tip:")
display(Audio("E-audio.mp3"))
print("Korean [a]:")
display(Audio("K-audio.mp3"))
etts(text1)
print("English [ɑ]:")
display(Audio("E-audio.mp3"))
```

# A. Warming up: [ ɛ ]

```python
#

Positions = "1-At the beginning of words" #@param = ["1-At the beginning of words", "2-In the middle of words", "3-At the end of words"]

ps = Positions.split('-')
np = ps[0]
np = int(np) -1

wordlist = ["end, egg, else, effort, elephant, extra, every, error",
            "bed, next, west, bent, many, brea, yes, meant, measure",
            "There is no word ending with this vowel in English."]

wlist = wordlist[np]
intro = ps[1] + ":"
mytxt = intro + wlist + "." + "This is the end."
etts(mytxt)

Audio("E-audio.mp3", autoplay = True)

```
