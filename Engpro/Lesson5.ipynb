# Lesson 5. Vowel pair 4: [ɑ] as in ‘father' (cf. Korean '아')

## 📌 Setting ups

```python
#@markdown 💠 Import libraries {Image, Audio, YouTubeVideo}
from IPython.display import Image, Audio, YouTubeVideo
```

```python
#@markdown 🔄 Making functions: etts('text'), ktts('text'), btts('text)
%%capture
!pip install gTTS
from gtts import gTTS
from IPython.display import Audio

def etts(text):
  text_to_say = text

  gtts_object = gTTS(text = text_to_say,
                     lang = "en", tld = "us",
                    slow = True)

  gtts_object.save("E-audio.mp3")
  return Audio("E-audio.mp3")

def ktts(text):
  text_to_say = text

  gtts_object = gTTS(text = text_to_say,
                     lang = "ko",
                    slow = False)

  gtts_object.save("K-audio.mp3")
  return Audio("K-audio.mp3")

def btts(text):
  text_to_say = text

  gtts_object = gTTS(text = text_to_say,
                     lang = "en", tld = "co.uk",
                    slow = True)

  gtts_object.save("B-audio.mp3")
  return Audio("B-audio.mp3")
```

# [1] Korean vs. English vowels: [ a ] vs. [ ɑ ]

```python
#@markdown Korean accent:
KSpelling = "1. Spelling 'o'" #@param = ["1. Spelling 'o'","2. Spelling 'a'"]
Kpn = KSpelling.split('. ')
Kpn = int(Kpn[0]) -1

#@markdown English:

ESpelling = "1. Spelling 'o'" #@param = ["1. Spelling 'o'","2. Spelling 'a'"]
Epn = ESpelling.split('. ')
Epn = int(Epn[0]) -1

wordlist1 = ["캅, cod, lock, comedy, contrary, company, copy, 옥슨, option, on, odd, honest, 숍 또는 샵, rocket, block, cot, top, fox, 스포트라이트, 오페라, 팔로우, constitution",
         "father, arm, want, wallet, dark, wasp, watch, March"]

kwlist = wordlist1[Epn]

if Kpn == 0:
  ktxt = "제가 스펠링 'o'가 포함된 단어들을 말해보겠습니다. 몇 개의 다른 모음으로 말하는지 체크해 보세요. " + kwlist

elif Kpn == 1:
  txt = "제가 스펠링 'a'가 포함된 단어들을 말해보겠습니다. 몇 개의 다른 모음으로 말하는지 체크해 보세요. " + kwlist
ktts(ktxt)



wordlist2 = ["cop, cod, lock, comedy, contrary, company, copy, oxen, option, on, odd, honest, shop, rocket, block, cot, top, fox, spot, opera, follow, constitution",
         "father, arm, want, wallet, dark, wasp, watch, March"]

wlist = wordlist2[Kpn]

if Epn == 0:
  txt = "Let me say 'o' spelling included words. " + wlist

elif Epn == 1:
  txt = "Let me say 'A' spelling including words. " + wlist
etts(txt)

print("English")
print(wlist)
display(Audio("E-audio.mp3"))

print("Korean")
print(kwlist)
display(Audio("K-audio.mp3"))
```
