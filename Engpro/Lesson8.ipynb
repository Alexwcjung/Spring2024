```python
#@markdown ðŸ”„ Making functions: etts('text'), ktts('text'), btts('text)
%%capture
!pip install gTTS
from gtts import gTTS
from IPython.display import Audio

def etts(text):
  text_to_say = text

  gtts_object = gTTS(text = text_to_say,
                     lang = "en", tld = "us",
                    slow = False)

  gtts_object.save("E-audio.mp3")
  return Audio("E-audio.mp3")

def ktts(text):
  text_to_say = text

  gtts_object = gTTS(text = text_to_say,
                     lang = "ko",
                    slow = False)

  gtts_object.save("E-audio.mp3")
  return Audio("E-audio.mp3")

def btts(text):
  text_to_say = text

  gtts_object = gTTS(text = text_to_say,
                     lang = "en", tld = "co.uk",
                    slow = False)

  gtts_object.save("E-audio.mp3")
  return Audio("E-audio.mp3")

def ctts(text):
  text_to_say = text

  gtts_object = gTTS(text = text_to_say,
                     lang = "en", tld = "ca",
                    slow = False)

  gtts_object.save("E-audio.mp3")
  return Audio("E-audio.mp3")
```

```python
# Import the libraries.
from IPython.display import Audio
from gtts import gTTS
from pydub import AudioSegment
import io

positions = "1_word-initial" #@param = ["1_word-initial", "2_word-medial", "3_word-final"]
pos = positions.split("_")
num = pos[0]
n=int(num)-1

wordlist = ["ago, away, along, amaze, contain, suppose, asleep, upon, arrive, obtain",
            "agony, relative, holiday, buffalo, telephone, company, belief, jacket, cement, open, typical, cousin, capital, support, believe, signal, famous, nation, certain",
            "soda, sofa, zebra, papa"]
wd = wordlist[n]
intro2 = pos[1]
mytxt = "Let me say some English words appearing in %s. Listen how the schwa vowel is pronounced. "%intro2 + wd

# Function to create an audio file with silence between texts.
def generate_audio_with_silence(text_list, silence_duration=2000):
    combined_audio = AudioSegment.silent(duration=0)  # start with an empty AudioSegment

    for text in text_list:
        # Generate speech
        tts = gTTS(text=text, lang='en')
        fp = io.BytesIO()
        tts.write_to_fp(fp)
        # Move to the start of the BytesIO object
        fp.seek(0)
        audio = AudioSegment.from_file(fp, format="mp3")

        # Add audio and silence to the combined audio
        combined_audio += audio + AudioSegment.silent(duration=silence_duration)

    # Export combined audio
    combined_audio.export("/content/C_audio.mp3", format="mp3")
    return Audio("/content/C_audio.mp3", autoplay=False)

# Example usage of the function.
# text1 = f"Here's a list of example words in {intro2} position."
text1 = intro2
text2 = "Here's a list of example words in this position."
text3 = wordlist[n]
generate_audio_with_silence([text1, text2, text3])
```

# B. Schwa vowel can be written in different spellings "a, e, i, o, u"

```python
#@markdown Wordlist (1 seconds of silence between)
from gtts import gTTS
from pydub import AudioSegment
from pydub.playback import play
import io

# Your input text
txt = "arrive, oven, liquid, occur, upon, ashamed, belief, typical, lesson, column"

# Split the text into words
words = txt.split(', ')

# Create an empty audio segment for combining everything
combined = AudioSegment.empty()

# Duration of silence in milliseconds (0.5 seconds)
silence_duration = 1000

# Generate audio for each word and add silence
for word in words:
    # Use gTTS to synthesize audio from the text
    tts = gTTS(text=word, lang='en')
    # Save the synthesized audio to a buffer
    mp3_fp = io.BytesIO()
    tts.write_to_fp(mp3_fp)
    mp3_fp.seek(0)
    # Load this audio as an AudioSegment
    word_audio = AudioSegment.from_file(mp3_fp, format="mp3")
    # Add the audio segment to the combined audio
    combined += word_audio
    # Add silence after each word, except the last one
    if word != words[-1]:
        combined += AudioSegment.silent(duration=silence_duration)

# Export the combined audio to a file
combined.export("/content/combined_audio.mp3", format="mp3")

# If you want to play the audio directly in Colab, uncomment the following line:
Audio("/content/combined_audio.mp3")
```
