# **[1] _What do you think an accent is?_**

```python
#@markdown English accent: Type text
Spelling = "1" #@param = [1, 2, 3, 4, 5]
Language = "ko" #@param = ["en","fr","zh-CN","es", "ko", "ja"]
Dialect = "com.au" #@param = ["com.au","co.uk","us","ca","co.in","ie", "co.za"]
pn = int(Spelling) -1

words = input("Type text: ")

wlist = words
txt = "I'm learning English like you. Let's practice together. " + " " + wlist

if Language == "en":
  def tts(txt):
    text_to_say = txt
    gtts_object = gTTS(text = text_to_say,
                     lang = "en", tld = Dialect,
                    slow = False)
    gtts_object.save("E-audio.mp3")
    return Audio("E-audio.mp3")
else:
  def tts(txt):
    text_to_say = txt
    gtts_object = gTTS(text = text_to_say,
                       lang = Language, slow = False)
    gtts_object.save("E-audio.mp3")
    return Audio("E-audio.mp3")

tts(txt)
Audio("E-audio.mp3", autoplay = True)
```

# **[2] Understanding accents**

```python
#@markdown tts(txt) {English}: In addition to word stress,...
def tts(txt):
    text_to_say = txt
    gtts_object = gTTS(text = text_to_say,
                     lang = "en", tld = "us",
                    slow = False)
    gtts_object.save("E-audio.mp3")
    return Audio("E-audio.mp3")

text = """
In addition to word stress, which is assigned to a syllable, English words can receive accents within a sentence. This results in greater emphasis when stress and accent are combined. For example, stressed syllables are more emphasized than unstressed ones, and stressed syllables in accented words are stronger and longer than those in unaccented words.
"""

tts(text)
print(text)
display(Audio("E-audio.mp3"))

text="My younger brother ate ten chocolate cookies"
tts(text)
print(text)
display(Audio("E-audio.mp3"))
```
